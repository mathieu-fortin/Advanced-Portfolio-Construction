## Questions 1-6

```python
import statsmodels.api as sm

beer_excess = ind.loc["1991":"2018",['Beer']] - fff.loc["1991":"2018", ['RF']].values
mkt_excess = fff.loc["1991":"2018",['Mkt-RF']]
exp_var = mkt_excess.copy()
exp_var["Constant"] = 1
lm = sm.OLS(beer_excess, exp_var).fit()
lm.summary()
```

```python
steel_excess = ind.loc["1991":"2018",['Steel']] - fff.loc["1991":"2018", ['RF']].values
lm = sm.OLS(steel_excess, exp_var).fit()
lm.summary()
```

```python
lm = sm.OLS(beer_excess["2013":], exp_var["2013":]).fit()
lm.summary()
```

```python
lm = sm.OLS(steel_excess["2013":], exp_var["2013":]).fit()
lm.summary()
```

```python
ind["1991":"1993"].subtract(fff.loc["1991":"1993"]['RF'], axis='index').tail()
```

```python
mkt_excess = fff.loc["1991":"1993",['Mkt-RF']]
ind_excess = ind["1991":"1993"].subtract(fff.loc["1991":"1993"]['RF'], axis='index')
exp_var = mkt_excess.copy()
exp_var['Constant'] = 1.0
lms, betas = {}, {}
for col in ind_excess.columns:
    lms[col] = sm.OLS(ind_excess.loc[:, [col]], exp_var).fit()
    betas[col] = lms[col].params['Mkt-RF']
```

```python
sector = max(betas, key=lambda x: float(betas[x]))
print("Highest beta {:.2f} for {}".format(betas[sector], sector))
```

```python
sector = min(betas, key=lambda x: float(betas[x]))
print("Lowest beta {:.2f} for {}".format(betas[sector], sector))
```

## Questions 7-10

```python
ind_excess = ind["1991":"2018"].subtract(fff.loc["1991":"2018"]['RF'], axis='index')
fama = fff.loc["1991":"2018"].copy()
fama['Constant'] = 1.0
lms, betas = {}, {}
for col in ind_excess.columns:
    lms[col] = sm.OLS(ind_excess.loc[:, [col]], fama).fit()
    betas[col] = lms[col].params
```

```python
sector = max(betas, key=lambda x: betas[x]['SMB'])
print("Highest SmallCap tilt {:.2f} for {}".format(betas[sector]['SMB'], sector))
```

```python
sector = min(betas, key=lambda x: betas[x]['SMB'])
print("Highest LargeCap tilt {:.2f} for {}".format(betas[sector]['SMB'], sector))
```

```python
sector = max(betas, key=lambda x: betas[x]['HML'])
print("Highest Value tilt {:.2f} for {}".format(betas[sector]['HML'], sector))
```

```python
sector = min(betas, key=lambda x: betas[x]['HML'])
print("Highest Growth tilt {:.2f} for {}".format(betas[sector]['HML'], sector))
```
